### Model for the allosteric regulation of metabolic enzymes

The PFK rate expression $\hat{r}_{1} = r_{1}v\left(\dots\right)_{1}$ is the product of a kinetic limit:

$$ r_{1} = k_{cat}E_{1}\left(\frac{F6P}{K_{F6P}+F6P}\right)\left(\frac{ATP}{K_{ATP}+ATP}\right) $$

and a control _correction_ of the form:

$$ v\left(\dots\right)_{j} = \frac{\displaystyle{\sum_{i\in\left\{X\right\}}W_{i}f_{i}\left(\dots\right)}}{\displaystyle{\sum_{j\in\mathcal{C}_{j}}W_{j}f_{j}\left(\dots\right)}} $$

where $W_{i}$ (dimensionless) denotes the weight of configuration $i$, while $f_{i}\left(\cdots\right)$ (dimensionless) is a hill-binding function $f_{i} = (x/K_{i})^{n_{i}}/(1+(x/K_{i})^{n_{i}})$
  which describes the fraction of bound activator/inhibitor ($x$) for configuration $i$; $K_{i}$ denotes a
  binding constant (mM), and $n_{j}$ denotes an order parameter (dimensionless).

#### a) Estimate the W's in the control correction function

The kinetic limit is constant for this problem as the two substrate F6P and ATP are constant in the experiment. On the other hand, the concentration of the activator (3-5-AMP) is increased over the course of the experiment. Looking at the data, we have activity when 3-5-AMP is zero, thus we'll have two W's in the control expression: let $W_{1}$ denote the activity correction in the absence of 3-5-AMP and $W_{2}$ will be the weight for the 3-5-AMP

##### Estimate $W_{1}$: 
$W_{1} = \frac{d_{1}}{r_{1}}$, where $d_{1}$ denotes the rate at 3-5-AMP = 0 (first data point) and $r_{1}$ denotes the kinetic limit. __This gives a $W_{1}$ = 0.044__

##### Estimate $W_{2}$:
At the other extreme, from the data we know the rate saturates at 3-5-AMP becomes large. Thus, lets assume at the last data point (say $N$) that the binding function is saturated; we can then solve for $W_{2}$ directly:

$$ W_{2} = \frac{(W_{1}-(1+W_{1})\alpha)}{(\alpha - 1)} $$

where $\alpha = d_{N}/r_{1}$. __This gives the value $W_{2}$ = 74.028__

#### b) Estimate binding function parameters $n$ and $K$:

I estimated these parameters by trial and error: We made the assumption when estimating $W_{2}$ that we were saturated at the last 3-5-AMP data point i.e., f(1mM) = 1. There are several strategies to get at these parameters. A quick and dirty method is to pick a value for $K\simeq{0.4}$ (mM) and then estimate the $n$ parameter, $n\simeq{4}$. Since f=1, we know that $(x/K)^{n}\gg{1}$, I picked some some values that fit the data.__ NOTE: This answer is NOT unique because you could have picked another value for K, or had a different understanding of $\gg$. __
